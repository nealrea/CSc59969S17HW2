```python
%matplotlib inline
import numpy as np
import matplotlib


import matplotlib.pyplot as plt
import matplotlib as mpl

matplotlib.rcParams['image.cmap'] = 'viridis'
%matplotlib inline
matplotlib.style.use('ggplot')
matplotlib.rcParams['figure.figsize'] = (15, 8)

data = np.recfromcsv('crabs.csv')

#Problem 1:

plt.scatter(data.fl, data.bd)
plt.xlabel('Frontal Lobe Size (mm)', fontsize = 20)
plt.ylabel('Body Depth (mm)', fontsize = 20)
plt.title('Leptograpsus Variegatus Frontal Lobe Size vs Body Depth', fontsize = 20)
data.dtype.fields
```

```python
#Problem 2

for x in range(0, len(data.cl)):
    plt.scatter(data.fl[x], data.bd[x], s = data.cl[x]*2, c = 'tomato')
plt.ylim(ymin = 0)
plt.xlim(xmin = 0)
plt.figtext(0.5,0,"Carapace length is notated by marker size.",horizontalalignment = 'center',fontsize = 20)
plt.xlabel('Frontal Lobe Size (mm)', fontsize = 20)
plt.ylabel('Body Depth (mm)', fontsize = 20)
plt.title('Leptograpsus Variegatus Frontal Lobe Size vs Body Depth vs Carapace Length', fontsize = 20)
plt.show()

plt.scatter(data.fl, data.bd, c = data.cl)
plt.ylim(ymin = 0)
plt.xlim(xmin = 0)
plt.xlabel('Frontal Lobe Size (mm)', fontsize = 20)
plt.ylabel('Body Depth (mm)', fontsize = 20)
plt.title('Leptograpsus Variegatus Frontal Lobe Size vs Body Depth vs Carapace Length', fontsize = 20)

colormap = plt.cm.get_cmap()
plot = plt.scatter(data.fl, data.bd, c = data.cl, vmin = 0, vmax = 50, cmap = colormap)
cbar = plt.colorbar(plot)
cbar.ax.set_ylabel('Carapace Length (mm)', rotation = 270, va = 'bottom', fontsize = 20)
```

Both of these graphs show the direct proportionality between Frontal Lobe Size, Body Depth, and Carapace Length.  The first graph utilizes marker size to make the increase in Carapace Length more obvious than the second, since size is an ordinal attribute.  However, the it is easier to quantify the Carapace Length in the second graph by using the color scale on the side.

```python
#Problem 3

blueMaleFL = np.array(data.fl[0:50])
blueMaleBD = np.array(data.bd[0:50])
blueMaleCL = np.array(data.cl[0:50])

blueFemaleFL = np.array(data.fl[50:100])
blueFemaleBD = np.array(data.bd[50:100])
blueFemaleCL = np.array(data.cl[50:100])

orangeMaleFL = np.array(data.fl[100:150])
orangeMaleBD = np.array(data.bd[100:150])
orangeMaleCL = np.array(data.cl[100:150])

orangeFemaleFL = np.array(data.fl[150:200])
orangeFemaleBD = np.array(data.bd[150:200])
orangeFemaleCL = np.array(data.cl[150:200])

#plotting blue male data
for x in range(0, 50):
    plt.scatter(blueMaleFL[x], blueMaleBD[x], s = blueMaleCL[x]*2, marker = 'o', c = 'dodgerblue')
plt.plot(blueMaleFL, blueMaleBD, c = 'dodgerblue', label = 'Blue Male')

#plotting blue female data
for x in range(0, 50):
    plt.scatter(blueFemaleFL[x], blueFemaleBD[x], s = blueFemaleCL[x]*2, marker = 'o', c = 'cyan')
plt.plot(blueFemaleFL, blueFemaleBD, c = 'cyan', linestyle = 'dashed', label = 'Blue Female')

#plotting orange male data
for x in range(0, 50):
    plt.scatter(orangeMaleFL[x], orangeMaleBD[x], s = orangeMaleCL[x]*2, marker = 'o', c = 'orangered')
plt.plot(orangeMaleFL, orangeMaleBD, c = 'orangered', label = 'Orange Male')

#plotting orange female data
for x in range(0, 50):
    plt.scatter(orangeFemaleFL[x], orangeFemaleBD[x], s = orangeFemaleCL[x]*2, marker = 'o', c = 'orange')
plt.plot(orangeFemaleFL, orangeFemaleBD, c = 'orange', linestyle = 'dashed', label = 'Orange Female')

plt.xlabel('Frontal Lobe Size (mm)', fontsize = 20)
plt.ylabel('Body Depth (mm)', fontsize = 20)
plt.title('Leptograpsus Variegatus Frontal Lobe Size vs Body Depth vs Carapace Length', fontsize = 20)
plt.ylim(ymin = 0)
plt.xlim(xmin = 0)
plt.legend(loc = 'best', fontsize = 16)
```
