```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import pyplot as plt
from scipy.io import loadmat
from scipy import stats as stats

matplotlib.rcParams['image.cmap'] = 'viridis'
%matplotlib inline
matplotlib.style.use('ggplot')
matplotlib.rcParams['figure.figsize'] = (15, 8)

data = loadmat('AQUA.2013089.2055.mat')
```

```python
print(data.keys())
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
cmap = plt.get_cmap('hot')
fig_image = ax.imshow(data['B1'], cmap = cmap)
cax = fig.colorbar(fig_image, ax = ax)
cax.set_label("Reflectance")
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
cmap = plt.get_cmap('hot')
fig_image = ax.imshow(data['B6'], cmap = cmap)
cax = fig.colorbar(fig_image)
cax.set_label("Reflectance")
```

```python
print(data['B1'].shape)
print(data['B6'].shape)
```

```python
N = 914*667
spercentage = 0.0045
numSamps = N*spercentage
samplesInds = np.array(range(1,N))
np.random.shuffle(samplesInds)
samplesInds = samplesInds[0:numSamps]

b1 = data['B1']
b6 = data['B6']
plt.scatter(b1.ravel()[samplesInds], b6.ravel()[samplesInds], alpha = 0.2, s = 45, c = 'darkred', marker = 'D')
plt.xlabel('Band 1 Reflectance', fontsize = 16)
plt.ylabel('Band 6 Reflectance', fontsize = 16)
plt.title('A Comparison of Band 1 and Band 6 Reflectance', fontsize = 20)
```

```python
#Band 1 vs Band 6
coeff = np.polyfit(b1.ravel()[samplesInds], b6.ravel()[samplesInds], 1)
plt.scatter(b1.ravel()[samplesInds], b6.ravel()[samplesInds], alpha = 0.1, s = 45, c = 'darkred', marker = 'D')
plt.xlabel('Band 1 Reflectance', fontsize = 16)
plt.ylabel('Band 6 Reflectance', fontsize = 16)
plt.title('A Comparison of Band 1 and Band 6 Reflectance', fontsize = 20)
plt.plot([0.4,1.0], np.polyval(coeff, [0.4,1.0]), color = 'red', label = 'Linear Regression Line')
plt.legend(loc = 'best', fontsize = 16)

slope = format(((np.polyval(coeff, 1.0) - np.polyval(coeff, 0.4))/(1.0-0.4)), '.4f')
plt.text(0.99, 0.5, "slope = " + str(slope), horizontalalignment = 'center', fontsize = 16)

pcc = format(stats.pearsonr(b1.ravel()[samplesInds], b6.ravel()[samplesInds])[0], '.4f')
plt.text(0.906, 0.45, "Pearson Correlation Coefficient = " + str(pcc), horizontalalignment = 'center', fontsize = 16)
```

```python
#Band 7 vs Band 6
b7 = data['B7']
coeff = np.polyfit(b7.ravel()[samplesInds], b6.ravel()[samplesInds], 1)
plt.scatter(b7.ravel()[samplesInds], b6.ravel()[samplesInds], alpha = 0.05, s = 45, c = 'darkred', marker = 'D')
plt.xlabel('Band 7 Reflectance', fontsize = 16)
plt.ylabel('Band 6 Reflectance', fontsize = 16)
plt.title('A Comparison of Band 7 and Band 6 Reflectance', fontsize = 20)
plt.plot([0.02,0.45], np.polyval(coeff, [0.02,0.45]), color = 'red', label = 'Linear Regression Line')
plt.legend(loc = 'best', fontsize = 16)

slope = format(((np.polyval(coeff, 0.45) - np.polyval(coeff, 0.02))/(0.45-0.02)), '.4f')
plt.text(0.03, 0.5, "slope = " + str(slope), horizontalalignment = 'center', fontsize = 16)

pcc = format(stats.pearsonr(b7.ravel()[samplesInds], b6.ravel()[samplesInds])[0], '.4f')
plt.text(0.092, 0.45, "Pearson Correlation Coefficient = " + str(pcc), horizontalalignment = 'center', fontsize = 16)
np.polyval(coeff, [0.02,0.45])
```

We can plainly see that Band 7 and Band 6 are much more closely related than Band 1 and Band 6.
